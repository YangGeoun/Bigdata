# **데이터셋 통합**

```python
from datetime import datetime
```

```python
start_datetime = datetime.now()
# start_datetime = datetime.today.trftime("%Y-%m-%d %H:%M:%S")

df_bus_car_tot = pd.DataFrame()

for i in range(80):
    # 파일 불러오기
    file_path = f"./data_원본/trfcard ({i})/trfcard.csv"
    df_bus_card_org = pd.read_csv(file_path)
    file_path_col = f"./data_원본/trfcard ({i})/trfcard_columns.xlsx"
    df_col = pd.read_excel(file_path_col, header=2, usecols="B:C")

    # 컬럼명 한글로 바꾸기 위한 딕셔너리 생성 및 한글로 변경
    dic = dict()
    for i in range(len(df_col)):
        dic[df_col.iloc[i, 0]] = df_col.iloc[i, 1] 
    df_bus_card_kor = df_bus_card_org.rename(columns=lambda x : dic[x])
    
    # 승차시각, 하차시각 datetime 필드로 변경하기
    df_bus_card_kor = df_bus_card_kor.astype({"승차시각" : "str", "하차시각" : "str"})

    # 필요한 컬럼만 뽑아내기
    df_bus_card = df_bus_card_kor[["승차시각", "하차시각", "승객연령", "환승여부", "추가운임여부", "승차정류장", "하차정류장"]].copy()

    df_bus_card["승차시각"] = pd.to_datetime(df_bus_card.loc[:,"승차시각"])
    df_bus_card["하차시각"] = pd.to_datetime(df_bus_card.loc[:,"하차시각"])

    ## 필요한 추가 컬럼 생성하기
    df_bus_card["버스내체류시간(분)"] = round((df_bus_card["하차시각"] - df_bus_card["승차시각"]).dt.total_seconds() / 60, 2)
    df_bus_card["기준년도"] = df_bus_card["승차시각"].dt.year
    df_bus_card["기준월"] = df_bus_card["승차시각"].dt.month
    df_bus_card["기준일"] = df_bus_card["승차시각"].dt.day
    df_bus_card["기준시각"] = df_bus_card["승차시각"].dt.hour
    df_bus_card["기준시각(분)"] = df_bus_card["승차시각"].dt.minute

    # 통합 데이터프레임에 가져온 데이터 추가(통합하기)
    df_bus_car_tot = pd.concat([df_bus_car_tot, df_bus_card], ignore_index=True)

end_datetime = datetime.now()

print(f"전체 실행 시간 : {start_datetime} ~ {end_datetime}")
print(f"작업 시간 : {end_datetime - start_datetime}")
```
