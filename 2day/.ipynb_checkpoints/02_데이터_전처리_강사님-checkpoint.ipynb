```python
### 행단위 조건 처리 -> 분석에서는 조건처리라는 용어를 --> 필터링 이라고 칭합니다.
```

```python
### tip의 값들이 5를 초과하는 모든 데이터를 조회해 주세요.
# - 모든 데이터란, 필터링 조건에 맞는 모든 열과 모든 행을 의미합니다.
### - 데이터프레임에서 조건처리(필터링) 할 때는 if문을 사용하지 않고 필터링 합니다.
# - 데이터 단위로 자동 처리됩니다.(데이터프레임이 알아서 True/False를 처리 해줍니다.)
c = tip["tip"] > 5
c

tip[c == True]

### True 생략가능
tip[(tip["tip"] > 5) == True]
tip[(tip["tip"] > 5)]
```

```python
### 성별 데이터들 중에 남자이면서, 담배를 피우지 않는 데이터 전체를 조회해 주세요.
### 기본 문법에서는 and, or 논리연사자를 사용
### 데이터프레임에서는 &, | 논리연산자 기호를 사용
c = (tip["sex"] == "Male") & (tip["smoker"] == "No")
c

tip[c == True]
tip[c]
tip[(tip["sex"] == "Male") & (tip["smoker"] == "No")]
```

```python
### 요일이 Sun 또는 Sat일때 방문한 사용자의 모든 데이터 조회하기
tip[(tip["day"] == "Sun") | (tip["day"] == "Sat")]

### or 조건인 경우에는 isin() 함수를 사용할 수 있습니다.
# isin() : 포함되어 있으면 True를 반환하는 기능을 수행함
c = tip["day"].isin(["Sun", "Sat"])
c
tip[c]
tip[tip["day"].isin(["Sun", "Sat"])]
```

### 결측치 처리하기

```python
### 데이터 불러들이기
# - 파일 데이터 사용
# - 파일을 불러들이거나 처리할 때 사용하는 라이브러리 -> 판다스(pandas)
# - UnicodeDecodeError : 문서 타입이 맞지 않을 때 발생하는 오류로
#                      : 언어 타입이 맞지 않아서 발생함
#                      : 문서를 만들때의 언어 타입으로 맞춰주어야 합니다.
#                      : 기본적으로 사용하는 전세계 표준 언어 타입은 "utf8"
#                      : 오류가 발생하면 "euc-kr" 언어 타입 사용
#                      : 그래도 오류가 발생하면 다른 방법 찾기
file_path = "./data/bicycle.csv"
df = pd.read_csv(file_path, encoding="euc-kr")
df
```

```python
df.info()
```

```python
### 결측치 데이터 확인하기
# 결측치가 있는(True)인 컬럼의 데이터들은 True로 표시됨
df.isnull()

# 결측치가 없는(True)인 컬럼의 데이터들을 True로 표시됨
# df.notnull()
```

```python
"""
<결측치 처리 방법>
1. 결측치가 있는 부분의 데이터를 사용할지 또는 말지(삭제) 결정
2. 사용하지 안는다면, 컬럼과 행 중에 어느 부분을 삭제 할지 결정
3. 사용한다면, 어떻게 값을 넣을지 결정(대체 방법 찾기)
    - 해당 결측치가 숫자인 경우에는 해당 열(컬럼)의 평균으로 대체
    - 범주형 데이터(문자)인 경우에는 범주의 비율대비 대체
    - 무조건 0으로 대체
    - 결측 데이터가 속한 행의 컬럼들과 유사한 타 데이터들의 평균을 사용(숫자 데이터인 경우)
    - 결측 데이터가 속한 컬럼 값의 이전/이후 값의 평균값을 취함(숫자 데이터인 경우)
"""
```

```python
### 컬럼별 결측 데이터 갯수 확인하기
df.isnull().sum(0)
```

```python
### 행단위 결측 데이터 취합하기
df.isnull().sum(1)
```

```python
### 결측치(NaN, None, Null)가 있는 모든 행 삭제하기
df_drop = df.dropna(axis=0)
df_drop.info()
df_drop
```

```python
### 결측치가 있는 모든 컬럼 삭제하기
df_drop_col = df.dropna(axis=1)
df_drop_col.info()
```

```python
### 특정 컬럼을 기준으로 행단위 삭제하기
# 대여거치대 데이터에 결측치가 있으면 삭제...
# df.info()
df["대여거치대"].dropna(axis=0)
```

```python
### 이용거리 데이터 중에 결측치가 있으면 0으로 대체하기(채우기)
# df.info()
df["이용거리"].fillna(0)
```

```python
df["이용거리"].fillna("missing")
```

```python
### 이용거리의 결측데이터를 -> 이용거리의 전체 평균으로 결측데이터 채우기
df["이용거리"].fillna(df["이용거리"].mean())
```

```python
df.info()
```

```python
df["대여거치대"] = df["대여거치대"].fillna(df["대여거치대"].mean())
df["반납거치대"] = df["반납거치대"].fillna(df["반납거치대"].mean())
df["이용시간"]   = df["이용시간"].fillna(df["이용시간"].mean())
df["이용거리"]   = df["이용거리"].fillna(df["이용거리"].mean())
```

```python
df.info()
```
