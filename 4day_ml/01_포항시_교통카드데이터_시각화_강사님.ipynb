```python
### 기본 시각화 라이브러리
import matplotlib
import matplotlib.pyplot as plt

### 기본 시각화보다 고급스러운 스타일의 시각화 라이브러리
import seaborn as sns

### 시각화를 위한 font 지정(한글깨짐 방지)
from matplotlib import font_manager, rc

### 사용자의 OS(운영체제) 확인 가능한 라이브러리
import platform

### 한글 font 설정
plt.rc("font", family = "Malgun Gothic")

### 시각화 시 마이너스(-, 음수) 기호 깨짐 방지
plt.rcParams["axes.unicode_minus"] = False
```

### [버스 이용량 분석]

```python
### 1. 기준월 및 기준일자별 버스 이용량 분석 비교
# - 기준월 및 기준일자별 버스 이용량 count 집계하기
df_pivot = df_bus_card_tot.pivot_table(values = "승객연령",
                                        index  = "기준월",
                                        columns= "기준일",
                                        aggfunc= "count")
df_pivot
```

```python
### 집계 데이터에서 NaN 데이터를 0으로 -> 결측치 처리하기
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
### 히트맵 그래프로 시각화 하기
# - 히트맵 : 데이터 3개가 필요함.. 보통 피봇테이블로 데이터 집계 후 사용하면 편함

# 그래프 사이즈 정의하기(너비, 높이)
plt.figure(figsize=(20, 10))

# 그래프 대제목 넣기
plt.title("기준월 및 기준일자별 버스 이용량 비교 분석")

# 히트맵 그리기 : seaborn 라이브러리에 있습니다.
# - annot : 박스 안에 숫자 보여줄지(True)/말지(False) 정의
# - fmt : 박수 안에 숫자의 길이 정의(소숫점이 있는경우 자릿수 지정)
# - cmap : 색상맵(색상지도), 즉 어떤 색상 그룹을 사용할지 지정
#          (인터넷 상에 매우 많은 색상맵이 존재함)
sns.heatmap(df_pivot, annot=True, fmt=".0f", cmap="rocket_r")

# 그래프 보여주기(그래프 실행)
plt.show()
```

### 기준일 및 기준시간대별 버스 내 체류시간(분) 비교 분석

```python
df_pivot = df_bus_card_tot.pivot_table(values   ="버스내체류시간(분)",
                                        index   ="기준일",
                                        columns ="기준시간",
                                         aggfunc="mean")
df_pivot
```

```python
### NaN 데이터를 0으로 결측치 처리하기
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
### 히트맵 그래프로 시각화 하기
# - 히트맵 : 데이터 3개가 필요함.. 보통 피봇테이블로 데이터 집계 후 사용하면 편함

# 그래프 사이즈 정의하기(너비, 높이)
plt.figure(figsize=(20, 10))

# 그래프 대제목 넣기
plt.title("기준일 및 기준시간별 버스 내 체류시간 비교 분석")

# 히트맵 그리기 : seaborn 라이브러리에 있습니다.
# - annot : 박스 안에 숫자 보여줄지(True)/말지(False) 정의
# - fmt : 박수 안에 숫자의 길이 정의(소숫점이 있는경우 자릿수 지정)
# - cmap : 색상맵(색상지도), 즉 어떤 색상 그룹을 사용할지 지정
#          (인터넷 상에 매우 많은 색상맵이 존재함)
sns.heatmap(df_pivot, annot=True, fmt=".0f", cmap="rocket_r")

# 그래프 보여주기(그래프 실행)
plt.show()
```

### 시간대별 승객연령별 버스내 체류시간(분) 분석

```python
### x축 : 시간
### y축 : 체류시간(분)
### 막대 : 승객연령별(일반, 청소년, ...) -> 범주별로
# -- 막대그래프 그리기
#   : 막대그래프를 그리기 위해서는 각 x축, y축, 막대범주에 대한 데이터를 
#     별도 데이터프레임으로 담아서 사용
```

```python
### 3개의 값을 담을 데이터프레임 정의하기
df_temp = pd.DataFrame()

### x축에 사용할 컬럼 생성
df_temp["기준시간"] = df_bus_card_tot["기준시간"]

### y축에 사용할 컬럼 생성
df_temp["버스내체류시간"] = df_bus_card_tot["버스내체류시간(분)"]

### 막대 범주로 사용할 컬럼 생성
df_temp["승객연령"] = df_bus_card_tot["승객연령"]
```

```python
df_temp.head(1)
```

```python
### 승객연령의 범주 데이터 확인하기
df_temp["승객연령"].unique()
```

```python
### 승객연령의 범주 데이터별 비율(빈도) 확인하기
df_temp["승객연령"].value_counts()
```

```python
### 시간대별 승객연령별 버스내 체류시간에 대해서 집계하기...
# - 이때 집계 데이터는 피봇테이블이 아닌 -> 그룹함수 사용
# - as_index=False : 그룹 단위로 표현되는 방식을 -> 행렬 단위의 데이터프레임 형태로 변환
df_temp2 = (df_temp.groupby(["기준시간", "승객연령"], 
                             as_index=False).sum()).sort_values(by=["버스내체류시간"], 
                                                                ascending=False)
df_temp2.head(1)
```

```python
### 막대그래프 그리기
# 그래프 너비, 높이 지정
plt.figure(figsize=(25, 10))

# 그래프 전체 제목
plt.title("시간대별 승객연령별 버스내 체류시간(분) 비교 분석")

# 막대그래프 그리기
# - hue : 범주 데이터를 가지는 컬럼 지정
sns.barplot(x="기준시간", y="버스내체류시간", hue="승객연령", data=df_temp2)

# 그래프 보여주기
plt.show()
```

```python
### 시간대별 승객연령 빈도(비율) 그래프 그리기
# - 히스토그램 그래프 사용

### 그래프 너비와 높이 지정
plt.figure(figsize=(12, 4))
### 그래프 전체 제목
plt.title("시간대별 승객연령 빈도(비율)분석")

### 히스토그램 그래프 그리기
sns.histplot(data = df_temp2,
             x = "기준시간",
             # 사용할 막대의 갯수
             bins = 30,
             # 막대그래프에 선그리기(밀도-빈도 선)
             kde = True,
             # 사용할 범주 데이터
             hue = "승객연령",
             # 각 범주를 쌓아 올리는 방식으로 표현
             multiple="stack",
             # 막대를 비율(밀도)로 표시(y축)
             stat = "density",
             # 막대의 너비 조절(0.6은 60%)
             shrink=0.5)

### 그래프 그리기
plt.show()
```

### [승하차정류장별 버스내 체류시간]
- 승하차정류장 구간별 버스내 체류시간(분)이 가장 높은 상위 30건 분석 비교

```python
df_bus_card_tot.head()
```

```python
### 승차정류장 데이터와 하차정류장 데이터를 합하여 구간 만들기
# 원본 데이터에서 가져올 컬럼 : 승차정류장+하차정류장, 버스내체류시간(분)
# 새로 만들 데이터 컬럼 : 승하차정류장, 버스내체류시간
df_temp3 = pd.DataFrame()

df_temp3["승하차정류장"] = df_bus_card_tot["승차정류장"] + "-->" + df_bus_card_tot["하차정류장"]
df_temp3["버스내체류시간"] = df_bus_card_tot["버스내체류시간(분)"]

df_temp3
```

```python
### 승하차정류장 별로 그룹을 지어서 버스내체류시간을 합한 후
#   - 버스내체류시간을 기준으로 내림차순 정렬하기
# ** 최종 결과를 담을 변수 : df_temp_sel

df_temp_sel = (df_temp3.groupby(["승하차정류장"], 
                              as_index=False).sum()).sort_values(by=["버스내체류시간"],
                                                                 ascending=False)

df_temp_sel_30 = df_temp_sel.head(30)
df_temp_sel_30
```

```python
### 선그래프 그리기
# 승하차정류장별 버스내체류시간 상위 30건 분석비교

# 그래프 사이즈 정의
plt.figure(figsize=(12, 4))
# 그래프 전체 제목 정의
plt.title("승하차정류장별 버스내체류시간 상위 30건 분석비교")

### 선 그리기
# x축 : 승하차정류장
# y축 : 버스내체류시간
plt.plot(df_temp_sel_30["승하차정류장"], df_temp_sel_30["버스내체류시간"])

# x축 제목
plt.xlabel("승하차정류장")
# y축 제목
plt.ylabel("버스내체류시간")

# x축 표시값의 각도 조절하기
plt.xticks(rotation=90)

# 그래프 내에 그리드 선 표시하기
plt.grid(True)

# 그래프 이미지 파일로 저장하기
# - show() 하기 전에 저장해야 합니다.
save_path = "./image/img01.png"
plt.savefig(save_path)

# 그래프 그리기
plt.show()
```
